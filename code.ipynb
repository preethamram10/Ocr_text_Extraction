```python
import easyocr
import re

# Function to extract text using EasyOCR
def extract_text_easyocr(image_path):
    # Initialize the EasyOCR reader
    reader = easyocr.Reader(['en'])  # You can specify more languages if needed
    
    # Perform OCR on the image
    result = reader.readtext(image_path, detail=0)  # Set detail=0 for only text
    
    # Join the list of text results into a single string
    extracted_text = ' '.join(result)
    return extracted_text

# Function to extract potential brand names using regular expressions
def extract_brand_name(text):
    # Example list of common brand names for demonstration; you can expand this
    brand_keywords = [
        "Surf Excel", "Nike", "Adidas", "Apple", "Samsung", "Sony", 
        "Coca-Cola", "Pepsi", "Microsoft", "Google", "Unilever"
    ]
    
    # Create a regex pattern that looks for any of the above brand names
    brand_pattern = "|".join(brand_keywords)
    
    # Use regular expression to find a brand name match
    brand_match = re.search(brand_pattern, text, re.IGNORECASE)
    
    if brand_match:
        return brand_match.group()
    
    # If no brand from the list is found, try a broader pattern that looks for capitalized words
    capitalized_words = re.findall(r'\b[A-Z][a-zA-Z]*\b', text)
    
    if capitalized_words:
        return ", ".join(capitalized_words)  # Return all potential brand-like words
    else:
        return "No brand found."

# Main function to run the process
def main(image_path):
    # Step 1: Extract text from the image using EasyOCR
    extracted_text = extract_text_easyocr(image_path)
    
    # Print extracted text for debugging purposes
    print("Extracted Text:\n", extracted_text)
    
    # Step 2: Find and display the brand name
    brand_name = extract_brand_name(extracted_text)
    
    print("Detected Brand Name:", brand_name)

# Example usage:
if __name__ == "__main__":
    # Provide the path to your image containing the product
    image_path = r'C:\Users\dell\Desktop\Flipkart Grid\new label datesets\Screenshot (97).png'  # Update with your actual image path
    main(image_path)
```
